```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv(r'lc_trainingset.csv')
pd.set_option('display.max_columns', None)
train['issue_d'] = pd.to_datetime(train['issue_d'])
train['earliest_cr_line'] = pd.to_datetime(train['earliest_cr_line'])
train.head()
```

```python
train_new = train.sample(frac=0.01)
```

```python
null_var = train_new.isna().sum() / train_new.shape[0] * 100
null_var
```

```python
see = null_var[null_var>0]
see
```

```python
for i in see.items():
    print(train_new[i[0]].dtypes)
```

```python
num_var = train_new.select_dtypes(include=np.number).keys()
num_var
```

```python
for i in num_var:
    train_new[i].fillna(train_new[i].mean(), inplace=True)
```

```python
cat_var = train_new.select_dtypes(include='object').keys()
cat_var
```

```python
for i in cat_var:
    train_new[i].fillna(train_new[i].mode().iloc[0], inplace=True)
```

```python
train_new.isna().sum().sum()
```

```python
train_new['loan_status'].value_counts()
```

```python
train_new['loan_status'].replace({'Fully Paid':0, 'Charged Off':1}, inplace=True)
```

```python
import klib
klib.data_cleaning(train_new)
klib.convert_datatypes(train_new)
```

```python
train_new.info()
```

```python
train_new.corrwith(train_new['loan_status'])
```

```python
train_new.drop(['annual_inc', 'revol_bal', 'total_acc', 'mort_acc', 'address', 'title', 'purpose', 'issue_d'], axis=1, inplace=True)
```

```python
from pycaret.classification import *
```

```python
s = setup(data = train_new, target = 'loan_status', silent=True, imputation_type='iterative', numeric_imputation='mean', 
train_size=0.95, categorical_imputation='mode', numeric_iterative_imputer='et', categorical_iterative_imputer='et',
transformation=True, normalize=True, remove_multicollinearity=True, session_id=2022)
```

```python
compare_models()
```

```python
rf = create_model('rf')
```

```python
tuned_model = tune_model(rf, choose_better=True, optimize='Accuracy')
```

```python
test = pd.read_csv('cleantestdata.csv')
test.head()
```

```python
pred = predict_model(tuned_model, data = test)
```

```python
pred
```

```python
sub = pd.concat([pred['id'], pred['Score']], axis =1)
```

```python
sub
```

```python
sub.rename(columns={'id':'Id', 'Score':'Predicted'}, inplace=True)
```
